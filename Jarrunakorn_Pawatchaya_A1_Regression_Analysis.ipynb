# Importing Libraries

```python
# importing library
import pandas as pd # data science essentials
import matplotlib.pyplot as plt # essential graphical output
import seaborn as sns  # enhanced graphical output
import numpy as np # mathematical essentials
import statsmodels.formula.api as smf # regression modeling
from sklearn.model_selection import train_test_split # train/test split
from sklearn.linear_model import LinearRegression # linear regression
import sklearn.linear_model


# setting pandas print options
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)


# specifying file name
file = './birthweight_low3.xlsx'


# reading the file into Python
birthw = pd.read_excel(io =file,
                        header = 0,
                        sheet_name = 0)

##############################################################
### EXPLORING DATASET ###
##############################################################

# checking missing value
print(f"""{birthw.isnull().any(axis = 0)} """) 

# descriptive statistic of data set 
birthw.describe(include ='number')
```

# Observing Missing Values

```python
# observing missing value
birthw.isnull().sum(axis = 0)
```

```python
# flagging missing values

# looping to detect features with missing values
for col in birthw:

    # creating columns with 1s if missing and 0 if not
    if birthw[col].isnull().astype(int).sum() > 0:
        birthw['m_'+col] = birthw[col].isnull().astype(int)

        # summing the missing value flags to check the results of the loop above
birthw[    ['m_meduc', 'm_npvis',
             'm_feduc']    ].sum(axis = 0)
```

```python
# creating a dropped dataset to visualize 'meduc'
df_dropped = birthw.dropna()


# displaying the plot for 'Mas Vnr Area'
sns.histplot(x = 'meduc',
            data = df_dropped,
            kde = True)


# title and labels
plt.title('Distribution of Father Education')


# displaying the plot
plt.show()

# creating a dropped dataset to visualize 'npvis'
df_dropped = birthw.dropna()


# displaying the plot for 'Mas Vnr Area'
sns.histplot(x = 'npvis',
            data = df_dropped,
            kde = True)


# title and labels
plt.title('Distribution of Father Education')


# displaying the plot
plt.show()

# creating a dropped dataset to visualize 'feduc'
df_dropped = birthw.dropna()


# displaying the plot for 'Mas Vnr Area'
sns.histplot(x = 'feduc',
            data = df_dropped,
            kde = True)


# title and labels
plt.title('Distribution of Father Education')


# displaying the plot
plt.show()
```

```python
#finding the mode for columns with missing values
print(f""" 
Mode of Mother Education       : {birthw['meduc'].mode()}

Mode of Father Education       : {birthw['feduc'].mode()}

Mode of Number Pretanal Visits : {birthw['npvis'].mode()}
""")
```

```python
### imputing missing value  ###

### meduc ###

# instantiating an imputation value
fill1 = 12

# imputing 'meduc'
birthw['meduc'] = birthw['meduc'].fillna(fill1)

### npvis ###

# instantiating an imputation value
fill2 = 12

# imputing 'npvis'
birthw['npvis'] = birthw['npvis'].fillna(fill2)

### feduc ###

# instantiating an imputation value
fill3 = 12

# imputing 'meduc'
birthw['feduc'] = birthw['feduc'].fillna(fill3)
```

```python
# Pearson correlation

### CONTINUOUS VARIABLE CORRELATION ###

# creating a list of continuous features (including bwght)
continuous_data = ["bwght","mage","fage"]


# developing a correlation matrix based on continuous features
birthw_corr_con = birthw[continuous_data].corr(method = 'pearson')


# filtering the results to only show correlations with Sale_Price
birthw_corr_con.loc[ : , "bwght"].round(decimals = 2).sort_values(ascending = False)

### INTERVAL VARIABLES CORRELATION ###

# creating a list of continuous features (including bwght)
interval_data = ["bwght","monpre","npvis", "cigs", "drink"]


# developing a correlation matrix based on continuous features
birthw_corr_in = birthw[interval_data].corr(method = 'pearson')


# filtering the results to only show correlations with Sale_Price
birthw_corr_in.loc[ : , "bwght"].round(decimals = 2).sort_values(ascending = False)

### CATEGORICAL VARIABLES CORRELATION ###

# creating a list of continuous features (including bwght)
cat_data = ["bwght", "omaps", "fmaps","meduc", "feduc", "male", "mwhte", "mblck", "moth", "fwhte", "fblck", "foth"]


# developing a correlation matrix based on continuous features
birthw_corr_cat = birthw[cat_data].corr(method = 'pearson')


# filtering the results to only show correlations with Sale_Price
birthw_corr_cat.loc[ : , "bwght"].round(decimals = 2).sort_values(ascending = False)

print(f""" 

CONTINUOUS 
---------------
{birthw_corr_con.loc[ : , "bwght"].round(decimals = 2).sort_values(ascending = False)}

INTERVAL/COUNT
---------------
{birthw_corr_in.loc[ : , "bwght"].round(decimals = 2).sort_values(ascending = False)}

CATEGORICAL/OTHER
---------------
{birthw_corr_cat.loc[ : , "bwght"].round(decimals = 2).sort_values(ascending = False)}

""")
```

```python
# building a base model with a log transformed response variable

# INSTANTIATING a model type
lm_bwght_mage = smf.ols(formula = """log_bwght ~ log_mage""",
                        data = birthw)


# telling Python to FIT the data to the blueprint
results = lm_bwght_mage.fit()


# printing a summary of the results
print(results.summary())

# INSTANTIATING a model type
lm_bwght_fage = smf.ols(formula = """log_bwght ~ log_fage""",
                        data = birthw)


# telling Python to FIT the data to the blueprint
results = lm_bwght_fage.fit()


# printing a summary of the results
print(results.summary())

### ANALYSIS ###
# fage has a slightly higher/ around the same negative correlation of -0.54, while mage has -0.53
```

# Ordinary Least Squares Regression

```python
# declaring set of x-variables
x_variables = ['monpre','fage', 'cigs','drink','mwhte', 'mblck', 'moth',
               'fblck', 'log_fage', 'log_monpre', 'smoke', 'drinking']


# looping to make x-variables suitable for statsmodels
for val in x_variables:
    print(f"{val} +")
```

```python
# applying modelin scikit-learn

# preparing x-variables from the OLS model
ols_data = birthw.loc[ : , x_variables]


# preparing response variable
birthw_target = birthw.loc[ : , 'bwght']


###############################################
## setting up more than one train-test split ##
###############################################

# FULL X-dataset (normal Y)
x_train_FULL, x_test_FULL, y_train_FULL, y_test_FULL = train_test_split(
            birthw_data,     # x-variables
            birthw_target,   # y-variable
            test_size = 0.25,
            random_state = 219)


# OLS p-value x-dataset (normal Y)
x_train_OLS, x_test_OLS, y_train_OLS, y_test_OLS = train_test_split(
            ols_data,         # x-variables
            birthw_target,   # y-variable
            test_size = 0.25,
            random_state = 219)
```

```python
# INSTANTIATING a model object
lr = LinearRegression()


# FITTING to the training data
lr_fit = lr.fit(x_train_OLS, y_train_OLS)


# PREDICTING on new data
lr_pred = lr_fit.predict(x_train_OLS)


# SCORING the results
print('OLS Training Score :', lr.score(x_train_OLS, y_train_OLS).round(4))  # using R-square
print('OLS Testing Score  :',  lr.score(x_test_OLS, y_test_OLS).round(4)) # using R-square

lr_train_score = lr.score(x_train_OLS, y_train_OLS).round(4)
lr_test_score = lr.score(x_test_OLS, y_test_OLS).round(4)

# displaying and saving the gap between training and testing - in absolute value
print('OLS Train-Test Gap :', abs(lr_train_score - lr_test_score).round(4))
lr_test_gap = abs(lr_train_score - lr_test_score).round(4)

# testing score slightly higher than training 
```

```python
# zipping each feature name to its coefficient
lr_model_values = zip(birthw_data[x_variables].columns,
                      lr_fit.coef_.round(decimals = 2))


# setting up a placeholder list to store model features
lr_model_lst = [('intercept', lr_fit.intercept_.round(decimals = 2))]


# printing out each feature-coefficient pair one by one
for val in lr_model_values:
    lr_model_lst.append(val)
    

# checking the results
for pair in lr_model_lst:
    print(pair)
```
